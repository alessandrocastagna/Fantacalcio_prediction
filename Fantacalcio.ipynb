```python
import numpy as np
def replace_zero_with_mean(row):
    row_non_zero = row[row != 0]
    if len(row_non_zero) > 0:
        row[row <=0] = np.mean(row_non_zero)
    row=round(row,3)
    return row

df = df.apply(replace_zero_with_mean, axis=1)

df
```

# Neural Network
##### Recurrent Neural Network (RNN) to forecast the average "Fanta voto" for the next serie a football season, we will use a Long Short-Term Memory (LSTM)

```python
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Prepara i dati di input e output
X=[]
y=[]
for i in range(len(df.columns)):
  if i<len(df.columns)-2:
    X.append(df.iloc[:,[i,i+1]].values.astype(np.float32))
    y.append(df.iloc[:,[i+2]].values.astype(np.float32))

# Converti le liste X e y direttamente in tensori TensorFlow
X = tf.convert_to_tensor(X, dtype=tf.float32)
y = tf.convert_to_tensor(y, dtype=tf.float32)

# Dividi gli indici dei dati in training e validation set
indices = np.arange(len(X))
indices_train, indices_val = train_test_split(indices, test_size=0.2, random_state=42)


# Seleziona i dati di addestramento e di convalida
X_train = tf.gather(X, indices_train)
y_train = tf.gather(y, indices_train)
X_val = tf.gather(X, indices_val)
y_val = tf.gather(y, indices_val)


# Crea il modello LSTM
model = Sequential()
model.add(LSTM(32, input_shape=X_train.shape[1:], activation='relu', return_sequences=True))
model.add(Dropout(0.1))
model.add(LSTM(32, input_shape=X_train.shape[1:], activation='relu', return_sequences=True))
model.add(Dense(16))
model.add(Dense(32))
model.add(Dense(8))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')

# Addestra il modello
model.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_val, y_val))
```

```python
# Prepara gli ultimi due valori delle colonne come input
X=[]
for i in range(len(df.columns)):
  if i<len(df.columns)-1:
    X.append(df.iloc[:,[i,i+1]].values.astype(np.float32))

# Converti le liste X e y direttamente in tensori TensorFlow
X = tf.convert_to_tensor(X, dtype=tf.float32)


# Fai la previsione utilizzando il modello
predicted_output = model.predict(X)
```

# Forecast indexes
#### Has been calculated an index based on the predicted value for the next year and the games played last season

```python
df_prev=df[['previsione']]
stagione8 = pd.read_csv("/content/gdrive/MyDrive/Fantacalcio/2022_23.csv", sep=";")
df_prev = df_prev.merge(stagione8[['Nome','Squadra','Pv']], on='Nome', how='left')

min_rating = df_prev['previsione'].min()
max_rating = df_prev['previsione'].max()
max_presence = df_prev['Pv'].max()

def preview_index(avg_rating, num_presence, weight_avg_rating=0.7, weight_presence=0.3, min_rating=min_rating, max_rating=max_rating, max_presence=max_presence):
    normalized_avg_rating = (avg_rating - min_rating) / (max_rating - min_rating)
    normalized_presence = num_presence / max_presence
    realistic_value_index = weight_avg_rating * normalized_avg_rating + weight_presence * normalized_presence
    return realistic_value_index


index={}
for i in df_prev.iterrows():
  index[i[1]['Nome']]=preview_index(i[1]['previsione'], i[1]['Pv'])

prev = pd.DataFrame()
prev['Nome']=index.keys()
prev['index']=index.values()

df_prev = df_prev.merge(prev, on='Nome', how='inner')
df_prev = df_prev.merge(lista[['Nome','R']], on='Nome', how='left')
df_prev
```
